# Proyecto de Construcción


<p style='text-align: justify;'> Peralta & Serpell (1991) Destaca en su investigación las etapas del proceso de trabajo para la realización de cualquier tipo de obra de construcción. </p>

<p style='text-align: justify;'>Solminiac & Thenoux (1998) Así mismo, menciona en su investigación las distintas etapas del proceso de obra, desde cómo se inicia hasta los documentos y permisos necesarios para su culminación. </p>



<img src="../figuras/indicePC.png" width="1000">


## Definición de un proyecto de construcción

- <p style='text-align: justify;'>Presencia de una necesidad. </p>

- <p style='text-align: justify;'>Análisis.  </p>

- <p style='text-align: justify;'>Determinación de objetivos (sociales, económicos, funcionales, rentables). </p>

- <p style='text-align: justify;'>Definición y evaluación de alternativas de solución (estudios de viabilidad, evaluación, financiación). </p>


## Diseño

- <p style='text-align: justify;'>Analizar del terreno donde se construirá el proyecto (topografía, geología, hidrología, condiciones generales).</p>

- <p style='text-align: justify;'>Diseño arquitectónico.</p>

- <p style='text-align: justify;'>Diseño estructural.v

- <p style='text-align: justify;'>Análisis de impacto ambiental. </p>

- <p style='text-align: justify;'>Diseño de las instalaciones.</p>

- <p style='text-align: justify;'>Redacción de documentos de licitación. </p>

- <p style='text-align: justify;'>Constructibilidad y mantenimiento.</p>


## Licitación

- <p style='text-align: justify;'>Solicitar ofertas y adjudicar. </p>


## Construcción

- <p style='text-align: justify;'>Definición de estrategia de gestión de la calidad.</p>

- <p style='text-align: justify;'>Obtención de permisos. </p>

- <p style='text-align: justify;'>Redacción de bases administrativas incluyendo contratos, plazos, costos y relación entre propietarios y empresa contratistas.</p>

- <p style='text-align: justify;'>Planificación y programación de los trabajos (fijación de plazos parciales y globales, asignación de recursos).</p> 

- <p style='text-align: justify;'>Establecimiento de la metodología de trabajo.</p>

- <p style='text-align: justify;'>Estudio de presupuesto y costes de la obra. </p>

- <p style='text-align: justify;'>Contrato de mano de obra. </p>

- <p style='text-align: justify;'>Compra de materiales y/o alquiler de equipos.</p>

- <p style='text-align: justify;'>Materialización de la obra.</p>

- <p style='text-align: justify;'>Control. </p>

- <p style='text-align: justify;'>Auditoria Ambiental.</p>


## Puesta en Marcha

- <p style='text-align: justify;'>Puesta en marcha.</p>

- <p style='text-align: justify;'>Controles mediante pruebas y ensayos para evaluar la calidad de la construcción. </p>

- <p style='text-align: justify;'>análisis detallado de todos los materiales de construcción y acabados. </p>

- <p style='text-align: justify;'>Pruebas de desempeño.</p>

- <p style='text-align: justify;'>Recepción ya sea preliminar, parcial o total. </p>  

- <p style='text-align: justify;'>Aprobación definitiva.</p>


## Operación y mantenimiento

- <p style='text-align: justify;'>Uso y mantenciones.</p>


## Abandono
